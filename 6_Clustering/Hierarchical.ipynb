## imports

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

## Generate hierarchical-friendly dataset

```python
np.random.seed(42)
X, y_true = make_blobs(n_samples=400, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=42)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], color='gray', s=30, alpha=0.7)
plt.title("Original Data (No cluster labels shown)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

## Plot dendrogram

```python
# Calculate linkage matrix using Ward's method (minimizes variance within clusters)
Z = linkage(X, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode='level', p=5)
plt.title("Hierarchical Clustering Dendrogram (truncated)")
plt.xlabel("Sample index")
plt.ylabel("Distance")
plt.show()
```

## Agglomerative Clustering and visualization

```python
agglo = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
labels = agglo.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', s=30)
plt.title("Hierarchical Agglomerative Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
